```python
class Solution(object):
    def missingNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        a=len(nums)
        a=(a*(a+1))//2
        for i in nums:
          a-=i
        return a
```

```python
a=Solution()
print(a.missingNumber([4,3,1,0]))
```
