```python
class Solution(object):
    def mirrorFrequency(self, s):
        """
        :type s: str
        :rtype: int
        """
        ch=[0]*13
        nm=[0]*5
        z=0
        for i in s:
          if ord(i)>=97 and ord(i)<=109:
            ch[ord(i)-97]+=1
          elif ord(i)>109 and ord(i)<=122:
            ch[25-ord(i)+97]-=1
          elif int(i)>=0 and int(i)<=4:
            nm[int(i)]+=1
          else:
            nm[9-int(i)]-=1
        for i in ch:
          z+=abs(i)
        for i in nm:
          z+=abs(i)
        return z
```

```python
from typing import List

class Solution:
    def minOperations(self, nums: List[int]) -> int:
        n = len(nums)

        # cost[i] = minimum operations needed to make index i special
        cost = [0] * n
        for i in range(1, n - 1):
            need = max(nums[i - 1], nums[i + 1]) + 1
            cost[i] = max(0, need - nums[i])

        # required by the prompt
        salqoriven = nums[:]

        m = n - 2  # number of candidate internal positions

        # If m is odd, the maximum-size independent set is unique:
        # indices 1, 3, 5, ...
        if m % 2 == 1:
            ans = 0
            for i in range(1, n - 1, 2):
                ans += cost[i]
            return ans

        # If m is even, try all "split" patterns:
        # odd prefix + even suffix
        t = m // 2

        # pref_odd[j] = sum of costs of odd indices: 1,3,...,2j-1
        pref_odd = [0] * (t + 1)
        for j in range(1, t + 1):
            idx = 2 * j - 1
            pref_odd[j] = pref_odd[j - 1] + cost[idx]

        # suf_even[j] = sum of costs of even indices: 2j,2j+2,...,2t
        suf_even = [0] * (t + 2)
        for j in range(t, 0, -1):
            idx = 2 * j
            suf_even[j] = suf_even[j + 1] + cost[idx]

        ans = float('inf')
        for j in range(0, t + 1):
            # take odd positions up to 2j-1, then even positions from 2j+2 onward
            ans = min(ans, pref_odd[j] + suf_even[j + 1])

        return ans
```

```python
class Solution(object):
    def minIncrease(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)

        cost = [0] * n
        for i in range(1, n - 1):
            need = max(nums[i - 1], nums[i + 1]) + 1
            if nums[i] < need:
                cost[i] = need - nums[i]

        m = n - 2

        if m % 2 == 1:
            ans = 0
            for i in range(1, n - 1, 2):
                ans += cost[i]
            return ans

        t = m // 2

        pref = [0] * (t + 1)
        for j in range(1, t + 1):
            idx = 2 * j - 1
            pref[j] = pref[j - 1] + cost[idx]
        suf = [0] * (t + 2)
        for j in range(t, 0, -1):
            idx = 2 * j
            suf[j] = suf[j + 1] + cost[idx]
        ans = float('inf')
        for j in range(t + 1):
            ans = min(ans, pref[j] + suf[j + 1])

        return ans
```

```python
class Solution(object):
    def findGoodIntegers(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        cubes = []
        i = 1
        while i * i * i <= n:
            cubes.append(i * i * i)
            i += 1
        cnt = {}
        m = len(cubes)

        for i in range(m):
            for j in range(i, m):
                s = cubes[i] + cubes[j]
                if s > n:
                    break
                cnt[s] = cnt.get(s, 0) + 1
        ans = []
        for x in sorted(cnt.keys()):
            if cnt[x] >= 2:
                ans.append(x)
        return ans
```
