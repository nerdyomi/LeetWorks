```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        area=0
        l=0
        r=len(height)-1
        while l<r:
          cur=min(height[l],height[r])*(r-l)
          area=max(area,cur)
          if height[l]<height[r]:
            l+=1
          else:
            r-=1
        return area
```

```python
a=Solution()
a.maxArea([1,8,6,2,5,4,8,3,7])
```
