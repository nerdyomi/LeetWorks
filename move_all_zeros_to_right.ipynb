```python
class Solution(object):
    def moveZeroes(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        def shiftleft(ar,i):
          while i<(len(ar)-1):
            ar[i]=ar[i+1]
            i+=1
          ar[i]=0
          return ar
        x=0
        for c in range(len(nums)):
          if nums[x]==0:
            nums=shiftleft(nums,x)
            x-=1
          x+=1
        return nums
```

```python
z=Solution()
print(z.moveZeroes([0,0,1]))
```
