```python
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        a=1
        b=2
        c=2
        if n==1:
            return a
        if n==2:
            return b
        while c<n:
            t=a+b
            a=b
            b=t
            c+=1
        return b
```

```python
a=Solution()
print(a.climbStairs(6))
```
