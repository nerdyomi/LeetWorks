```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        terget=None
        count=0
        for i in nums:
          if count==0:
            terget=i
          if terget==i:
            count+=1
          else:
            count-=1
        return terget
```

```python
a=Solution()
x=a.majorityElement([1,2,3,3,1,2,2,2])
print(x)
```
