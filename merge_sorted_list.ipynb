```python
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: None Do not return anything, modify nums1 in-place instead.
        """
        a=b=0
        r=[]
        while a<m and b<n:
          if nums1[a]<=nums2[b]:
            r.append(nums1[a])
            a+=1
          else:
            r.append(nums2[b])
            b+=1
        while a<m:
          r.append(nums1[a])
          a+=1
        while b<n:
          r.append(nums2[b])
          b+=1

        return r
```

```python
a=Solution()
print(a.merge([1,2,3,0,0,0],3,[2,5,6],3))
```
