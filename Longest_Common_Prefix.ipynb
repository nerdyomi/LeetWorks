```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        ans=""
        i=0
        if len(strs)==0:
          return ans
        if len(strs)==1:
          return strs[0]
        while i<len(strs[0]):
          st=strs[0][i]
          for j in range(1,len(strs)):
            if i==len(strs[j]) or strs[j][i]!=st:
              return ans
          ans+=st
          i+=1
        return ans
```
