```python
class Solution(object):
    def minCostClimbingStairs(self, cost):
        """
        :type cost: List[int]
        :rtype: int
        """
        if len(cost) == 2:
            return min(cost)
        if len(cost) == 3:
            return min(cost[0] + cost[2], cost[1])
        for i in range(2, len(cost)):
            cost[i]= min(cost[i-1] + cost[i], cost[i-2] + cost[i])
        return min(cost[-1],cost[-2])
```

```python
a=Solution()
a.minCostClimbingStairs([10,15,20])
```
