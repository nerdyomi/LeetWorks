```python
class Solution(object):
    def canCompleteCircuit(self, gas, cost):
        """
        :type gas: List[int]
        :type cost: List[int]
        :rtype: int
        """
        fuel=0
        total=0
        start=0
        for i in range(len(gas)):
            fuel+=gas[i]-cost[i]
            total+=gas[i]-cost[i]
            if fuel<0:
                fuel=0
                start=i+1
        if total>=0:
            return start
        else:
            return -1
```

```python
a=Solution()
print(a.canCompleteCircuit([1,2,3,4,5],[3,4,5,1,2]))
```
