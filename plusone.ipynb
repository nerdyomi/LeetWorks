```python
class Solution(object):
    def plusOne(self, digits):
        """
        :type digits: List[int]
        :rtype: List[int]
        """
        carry=1
        ind=len(digits)-1
        while carry and ind>=0:
          now=digits[ind]+carry
          digits[ind]=now%10
          carry=now//10
          ind-=1
        if carry:
          digits.insert(0,1)
        return digits
```

```python
a=Solution()
a.plusOne([2,2,9])
```
