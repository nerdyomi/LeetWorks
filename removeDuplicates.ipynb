```python
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        i = 0

        for j in range(1, len(nums)):
            if nums[j] != nums[i]:
                nums[i+1] = nums[j]
                i += 1

        return i+1
```

```python
a=Solution()
a.removeDuplicates([0,0,1,1,1,2,2,3,3,4])
```
