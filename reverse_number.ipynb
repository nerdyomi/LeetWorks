```python
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        a=str(x)
        if a[0]=="-":
            a=int("-"+a[-1:0:-1])
        else:
            a=int(a[-1::-1])
        if -2**31>a or a>2**31-1:
            return 0
        return a
```

```python
a=Solution()
print(a.reverse(-123))
```
