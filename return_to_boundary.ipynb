```python
class Solution(object):
    def returnToBoundaryCount(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        c=0
        s=0
        for i in nums:
            s+=i
            if s==0:
                c+=1
        return c
```

```python
a=Solution()
print(a.returnToBoundaryCount([1,0,-1,0,1]))
```
