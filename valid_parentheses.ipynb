```python
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        stack=[]
        mapping={')':'(','}':'{',']':'['}
        for i in s:
          if i in mapping.values():
            stack.append(i)
          elif i in mapping.keys():
            if stack and mapping[i]==stack[-1]:
              stack.pop()
            else:
              return False
        return not stack
```

```python
a=Solution()
a.isValid("]")
```

```python
a=Solution()
a.isValid("[(({}))]")
```
