```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        nums.sort()
        s=nums
        ans=[]
        print(s)
        i=0
        while i<len(nums)-2:
          j=i+1
          k=len(nums)-1
          while j<k:
            print(s[i]+s[j]+s[k],i,j,k,s)
            if s[i]+s[j]+s[k]==0:
              ans.append([s[i],s[j],s[k]])
              j+=1
              while s[j]==s[j-1] and j<k:
                j+=1
            elif s[i]+s[j]+s[k]<0:
              j+=1
              while s[j]==s[j-1] and j<k:
                j+=1
            else:
              k-=1
              while s[k]==s[k+1] and k>j+1:
                k-=1
          i+=1
          while s[i]==s[i-1] and i<len(s)-3:
            i+=1
          if s[i]==s[i-1]:
            break

        return ans
```
