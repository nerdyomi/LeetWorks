```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        profit=0
        bought=prices[0]
        for i in range(1,len(prices)):
          if prices[i-1]>=prices[i]:
            profit+=prices[i-1]-bought
            bought=prices[i]
        if prices[-1]>bought:
          profit+=prices[-1]-bought
        return profit
```

```python
a=Solution()
print(a.maxProfit([7,1,5,3,6,4]))
```
