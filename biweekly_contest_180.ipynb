```python
class Solution(object):
    def trafficSignal(self, timer):
        """
        :type timer: int
        :rtype: str
        """
        if timer==0:
          return "Green"
        elif timer==30:
          return "Orange"
        elif 30<timer<=90:
          return "Red"
        else:
          return "Invalid"
```

```python
class Solution(object):
    def countDigitOccurrences(self, nums, digit):
        """
        :type nums: List[int]
        :type digit: int
        :rtype: int
        """
        c=0
        for num in nums:
          while num>9:
            rem=num%10
            num=num//10
            if rem==digit:
              c+=1
          if num==digit:
            c+=1
        return c
```

```python
a=Solution()
print(a.countDigitOccurrences([],0))
```

```python
class Solution(object):
    def minOperations(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)

        M = 200000
        is_prime=[True]*(M+1)
        is_prime[0]=is_prime[1]=False

        for i in range(2, int(M**0.5) + 1):
            if is_prime[i]:
                for j in range(i*i,M+1,i):
                    is_prime[j]=False

        op=0
        for i in range(n):
            val=nums[i]
            if i%2==0:
                while not is_prime[val]:
                    val+=1
                    op+=1
            else:
                while is_prime[val]:
                    val+=1
                    op+=1
        return op
```

```python
a=Solution()
print(a.minOperations([1,2,3,4]))
```

```python
class Solution(object):
    def maxValue(self, nums1, nums0):
        """
        :type nums1: List[int]
        :type nums0: List[int]
        :rtype: int
        """
        MOD=10**9 + 7
        n=len(nums1)
        tup=[(nums1[i], nums0[i]) for i in range(n)]
        tup.sort(key=lambda x:(-x[0], -(x[0] + x[1])))
        res=0
        for a1,a0 in tup:
            length=a1+a0
            val=(pow(2,a1,MOD)-1)%MOD
            val=(val*pow(2,a0,MOD))%MOD
            res=(res*pow(2,length,MOD))%MOD
            res=(res+val)%MOD
        return res
```

```python
a=Solution()
print(a.maxValue([3,1],[0,3]))
```
