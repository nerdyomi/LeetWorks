```python
"""
# Definition for a Node.
class Node:
    def __init__(self, x, next=None, random=None):
        self.val = int(x)
        self.next = next
        self.random = random
"""

class Solution(object):
    def copyRandomList(self, head):
        """
        :type head: Node
        :rtype: Node
        """
        if head==None:
            return None
        dic={}
        cur=head
        while cur:
            dic[cur]=Node(cur.val)
            cur=cur.next

        cur=head
        while cur:
            if cur.next==None:
                dic[cur].next=None
            else:
                dic[cur].next=dic[cur.next]
            if cur.random==None:
                dic[cur].random=None
            else:
                dic[cur].random=dic[cur.random]
            cur=cur.next
        return dic[head]
```
