```python
class Solution(object):
    def isPowerOfTwo(self, n):
        """
        :type n: int
        :rtype: bool
        """
        if n<=0:
            return False
        b=bin(n)
        if b.count('1')==1:
            return True
        else:
            return False
```

```python
a=Solution()
print(a.isPowerOfTwo(3))
```
