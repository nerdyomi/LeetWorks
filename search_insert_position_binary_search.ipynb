```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        n = len(nums)
        l = 0
        h = n-1

        while l<=h:
            m = (l+h)//2

            if nums[m] >= target:
                h = m-1
            else:
                l = m+1

        return l
```

```python
a=Solution()
a.searchInsert([1,3,5,6], 5)
```
