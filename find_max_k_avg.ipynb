```python
class Solution(object):
    def findMaxAverage(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: float
        """
        m=sum(nums[:k])
        s=m
        i=0
        while i+k<len(nums):
          s=s-nums[i]+nums[i+k]
          if s>m:
            m=s
          i+=1
        return m/k
```

```python
a=Solution()
print(a.findMaxAverage([1,12,-5,-6,50,3],4))
```
